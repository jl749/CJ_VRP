```python
#지도의 처음 위치 찍기 -> 화성이랑 제일 가까운 CJ 물류 센터로 선택 
# map = folium.Map(location = [물류센터의 위도 , 물류센터의 경도],zoom_start=13)
map = folium.Map(location = [37.33507198212856, 126.7140465554204],zoom_start=13)


#물류센터 출발지 marker 로 표시
#start = folium.Marker(location = [물류센터의 위도 , 물류센터의 경도],popup="물류센터",icon=folium.Icon(color="red", icon="info-sign") ).add_to(map) 
start = folium.Marker(location = [37.33507198212856, 126.7140465554204],popup="물류센터",icon=folium.Icon(color="red", icon="info-sign") ).add_to(map)

#택배지 위도 경도 데이터를 지도에 시각화 
for ma in range(coordinates_data.shape[0]):
  folium.Marker(location = [coordinates_data['위도'][ma],coordinates_data['경도'][ma]],popup = '택배' + str(ma) + '주소').add_to(map)


map
```
