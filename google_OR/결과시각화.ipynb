```python
import folium
import pandas as pd 
import pickle5 as pickle
from typing import List
```

```python
with open("/content/adsWithCor.pkl","rb") as fr:
    data = pickle.load(fr)
print(data)


```

```python
route1 = [21,67,34,79,58,68,90,89] 
route2 = [93,30,27,47,97,26]
route3 = [13,85,39,76,87] 
route4 = [11,28,84,38,88,86,0,25,80,69,7,64,29,1,42,50,98,6,51,63,91,54,59,56,35,48,44]

all_route = []

all_route.append(route1)
all_route.append(route2)
all_route.append(route3)
all_route.append(route4)
```

```python
add_list = pd.read_csv('/content/sample_300.csv')
```

```python
add_list
```

```python
ad_list: List[str] = add_list['받는분주소'].sample(n=100, random_state=1).values
```

```python
ad_list
```

```python
#시작 지점 찍기 
x=37.13291249518938
y=127.03000384352325
m=folium.Map(location=[x,y],zoom_start=13)
m
```

```python
colors = ["red","green","blue","purple"]
folium.CircleMarker(location=[x,y],radius=10,popup="Start",
                    color = '#00FF00', fill_color = '#00FF00', fill=True).add_to(m)
for route,color in zip(all_route,colors):
  points =[([x,y])]
  for r in route:
    #배송할 주소지의 x,y 좌표
    addx,addy = data[r]
    points.append(tuple([addx,addy]))
    load_add = ad_list[r]
    folium.CircleMarker(location=[addx,addy],radius = 10,popup=load_add,color = '#00FF00', fill_color = '#00FF00', fill=True).add_to(m)
  
  folium.PolyLine(points, color=color, weight=2.5, opacity=1).add_to(m)
```
